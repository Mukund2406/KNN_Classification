```python
def create_piechart(data,column):
    """
    objective : Creat Piechart for cateorical variable present in the pamdas data frame 
    
    params :
            data =this is panda dataframe
            column: this is name ehich used to creeate plot 
    Retrun :
            this will show piechart 
    """        
    
    labels=list(data[column].value_counts().to_dict().keys())
    sizes = list(data[column].value_counts().to_dict().values())
    plt.pie(sizes,
            labels=labels,
            autopct="%1.2f%%",
            shadow= False,
            startangle=45)
    plt.axis("equal")
    plt.title("Piechart - {}".format(column))
    plt.show()
    
    
    
    
    
    
def check_missing_value(data):
    """
  objective: Check missing value count and persantage in all coloumn 
  
  data: input data frame 
  retrun: Missing value data frame
  
   """
    total =data.isnull().sum().sort_values(ascending=False)
    per=(data.isnull().sum()/data.isnull().count()*100).sort_values(ascending=False)
    missing_data= pd.concat([total,per],axis=1,keys=["Total","Percent"])
    return missing_data
```

## Data preprocesing 

```python
#feture selecation 

df_train= df_train[["Name","Sex","Pclass","Age","Survived"]]
df_test= df_test[["Name","Sex","Pclass","Age","Survived"]]
```

```python
## findding the missing value 

print(df_train.isnull().sum())
print()
print(df_test.isnull().sum())
    
```

```python
##inputing numarical missing value mean value 



df_train["Age"].fillna(df_train["Age"].mean() , inplace= True )


df_test["Age"].fillna(df_test["Age"].mean() , inplace= True )
```

```python
## findding the missing value 

print(df_train.isnull().sum())
print()
print(df_test.isnull().sum())
    
```

```python
df_train["Sex"]=df_train["Sex"].map({"male":0,"female":1})
df_test["Sex"]=df_test["Sex"].map({"male":0,"female":1})                                    
```

```python
df_train.head()
```

```python
X_train = df_train.values[:870,1:-1]
y_train = df_train.values[:870, -1]
y_train=y_train.astype(int)


X_test = df_test.values[:,1:-1]
y_test = df_test.values[:, -1]
y_test = y_test.astype(int)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X_train)

X_train_scale = scaler.transform(X_train)
X_test_scale = scaler.transform(X_test)
```

```python
X_test_scale
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
%%time

#create a model 
calssifier=LogisticRegression()

##fitting traing data to model 
calssifier.fit(X_train_scale,y_train)


# predict using the model 

y_pred=calssifier.predict(X_test_scale)

print(list(zip(y_test,y_pred)))

            
```

```python
print(list(zip(df_test.columns[:-1] , calssifier.coef_.ravel())))
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score , classification_report

cnf= confusion_matrix(y_test,y_pred)
print(cnf)

print("classification report:")

print(classification_report(y_test,y_pred))
acc=accuracy_score(y_test,y_pred)
print("accuracy of the model:",acc)
      
```

```python
y_pred_train=calssifier.predict(X_train_scale)

cnf= confusion_matrix(y_train,y_pred_train)
print(cnf)

print("classification report:")

print(classification_report(y_train,y_pred_train))
acc=accuracy_score(y_train,y_pred_train)
print("accuracy of the model:",acc)
      
```

```python
acc_dict = dict()

for k in range(1,31):
    # creat object with k value 
    model_KNN=KNeighborsClassifier(k, metric= "euclidean")
    
    # model fit 
    model_KNN.fit(X_train_scale , y_train)
    
    y_pred= model_KNN.predict(X_test_scale)
    
    acc_score = round(accuracy_score(y_test , y_pred),2)
    
    print("Accuracy is : {} for k value ; {}".format(acc_score,k))
    acc_dict[k] = acc_score
                                   
```

```python
for k in acc_dict:
    if acc_dict[k]==max(acc_dict.values()):
        print(k,":",acc_dict[k])
```

```python
model = KNeighborsClassifier(n_neighbors=13 , metric= "euclidean" , n_jobs=-1)


model.fit(X_train_scale , y_train)

y_pred=model.predict(X_test_scale)
```
